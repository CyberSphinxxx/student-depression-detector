## 1. Setup and Imports

```python
# First, we import the tools we need
import pandas as pd               # for working with tables/data
import numpy as np                # for math stuff
import matplotlib.pyplot as plt   # for making charts
import seaborn as sns             # for nicer-looking charts
import warnings
warnings.filterwarnings('ignore') # hide unnecessary warnings

# This just makes our charts look cleaner
sns.set_theme(style="whitegrid")
plt.rcParams["figure.dpi"] = 120

print("All libraries loaded!")
```

## 2. Dataset Overview

```python
# Load the CSV file into a table called 'df'
df = pd.read_csv("Student_Depression_Dataset.csv")

# Show the first 5 rows so we can see what the data looks like
print("First 5 rows of the dataset:")
df.head()
```

```python
# Check how many rows and columns we have
print("Shape (rows, columns):", df.shape)

# List all column names
print("\nColumn Names:")
for col in df.columns:
    print(" -", col)
```

```python
# Check what data type each column has
print("Data Types of Each Column:")
print(df.dtypes)
```

## 3. Data Quality Checks

```python
# Check for missing (empty) values in each column
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Check for duplicate rows (exact same row appearing more than once)
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

## 4. Class Distribution (Target Variable)

```python
# Count how many students are depressed (1) vs not depressed (0)
print("Depression column value counts:")
print(df["Depression"].value_counts())

print()

# Show the percentage too
total = len(df)
for val, count in df["Depression"].value_counts().items():
    label = "Depressed" if val == 1 else "Not Depressed"
    print(f"{label}: {count} students ({count/total*100:.1f}%)")
```

## 5. EDA Visualizations

### 5.1 Target Class Distribution

```python
# We want to see how many students are depressed vs not depressed
labels = ["Not Depressed", "Depressed"]
counts = df["Depression"].value_counts().sort_index().values
colors = ["#5BA4CF", "#E07B7B"]

# Create two charts side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Left chart: Bar chart ---
bars = axes[0].bar(labels, counts, color=colors, edgecolor="white", width=0.5)
axes[0].set_title("Class Distribution (Bar Chart)", fontweight="bold")
axes[0].set_ylabel("Number of Students")
axes[0].set_ylim(0, max(counts) * 1.15)

# Add the number on top of each bar
for bar, count in zip(bars, counts):
    axes[0].text(bar.get_x() + bar.get_width()/2,
                 bar.get_height() + 150,
                 f"{count:,}", ha="center", fontsize=11, fontweight="bold")

# --- Right chart: Pie chart ---
axes[1].pie(counts, labels=labels, autopct="%1.1f%%",
            colors=colors, startangle=90)
axes[1].set_title("Class Distribution (Pie Chart)", fontweight="bold")

plt.suptitle("Depression Status Among Students", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("viz1_class_distribution.png", bbox_inches="tight")
plt.show()
```

### 5.2 Age Distribution by Depression Status

```python
# Add a label column so charts are easier to read
df["Depression_Label"] = df["Depression"].map({0: "Not Depressed", 1: "Depressed"})

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: Histogram to see age spread ---
for val, label, color in [(0, "Not Depressed", "#5BA4CF"), (1, "Depressed", "#E07B7B")]:
    subset = df[df["Depression"] == val]["Age"]
    axes[0].hist(subset, bins=20, alpha=0.65, label=label, color=color, edgecolor="white")

axes[0].set_title("Age Distribution by Depression Status", fontweight="bold")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")
axes[0].legend()

# --- Right: Boxplot to compare age ranges ---
sns.boxplot(data=df, x="Depression_Label", y="Age",
            palette={"Not Depressed": "#5BA4CF", "Depressed": "#E07B7B"},
            ax=axes[1], linewidth=1.5)
axes[1].set_title("Age Boxplot by Depression Status", fontweight="bold")
axes[1].set_xlabel("")
axes[1].set_ylabel("Age")

plt.tight_layout()
plt.savefig("viz2_age_distribution.png", bbox_inches="tight")
plt.show()
```

### 5.3 Academic Pressure and CGPA

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: How depression changes with academic pressure level ---
sns.countplot(data=df, x="Academic Pressure", hue="Depression_Label",
              palette={"Not Depressed": "#5BA4CF", "Depressed": "#E07B7B"},
              ax=axes[0], edgecolor="white")
axes[0].set_title("Academic Pressure vs Depression", fontweight="bold")
axes[0].set_xlabel("Academic Pressure (1=Low, 5=High)")
axes[0].set_ylabel("Count")
axes[0].legend(title="Status")

# --- Right: CGPA distribution for depressed vs not ---
for val, label, color in [(0, "Not Depressed", "#5BA4CF"), (1, "Depressed", "#E07B7B")]:
    subset = df[df["Depression"] == val]["CGPA"]
    axes[1].hist(subset, bins=25, alpha=0.65, label=label, color=color, edgecolor="white")

axes[1].set_title("CGPA Distribution by Depression Status", fontweight="bold")
axes[1].set_xlabel("CGPA")
axes[1].set_ylabel("Frequency")
axes[1].legend()

plt.tight_layout()
plt.savefig("viz3_academic.png", bbox_inches="tight")
plt.show()
```

### 5.4 Sleep Duration and Dietary Habits

```python
# Define the order we want to display these categories
sleep_order = ["Less than 5 hours", "5-6 hours", "7-8 hours", "More than 8 hours", "Others"]
diet_order  = ["Unhealthy", "Moderate", "Healthy", "Others"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Left: Sleep duration vs depression ---
# Group and count, then reshape for easy plotting
sleep_data = df.groupby(["Sleep Duration", "Depression_Label"]).size().unstack(fill_value=0)
sleep_data = sleep_data.reindex([s for s in sleep_order if s in sleep_data.index])
sleep_data.plot(kind="bar", ax=axes[0], color=["#E07B7B", "#5BA4CF"], edgecolor="white", width=0.7)
axes[0].set_title("Sleep Duration vs Depression", fontweight="bold")
axes[0].set_xlabel("Sleep Duration")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis="x", rotation=25)
axes[0].legend()

# --- Right: Dietary habits vs depression ---
diet_data = df.groupby(["Dietary Habits", "Depression_Label"]).size().unstack(fill_value=0)
diet_data = diet_data.reindex([d for d in diet_order if d in diet_data.index])
diet_data.plot(kind="bar", ax=axes[1], color=["#E07B7B", "#5BA4CF"], edgecolor="white", width=0.7)
axes[1].set_title("Dietary Habits vs Depression", fontweight="bold")
axes[1].set_xlabel("Dietary Habits")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis="x", rotation=15)
axes[1].legend()

plt.tight_layout()
plt.savefig("viz4_sleep_diet.png", bbox_inches="tight")
plt.show()
```

### 5.5 Correlation Heatmap of Numeric Features

```python
# Pick only the number columns for correlation
num_cols = ["Age", "CGPA", "Academic Pressure", "Work Pressure",
            "Study Satisfaction", "Job Satisfaction",
            "Work/Study Hours", "Financial Stress", "Depression"]

# Calculate how related each pair of columns is (-1 to 1)
corr = df[num_cols].corr()

# Only show the bottom half of the table (the top half is the same)
mask = np.triu(np.ones_like(corr, dtype=bool))

fig, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f",
            cmap="RdYlBu_r", center=0, linewidths=0.5,
            ax=ax, annot_kws={"size": 9}, cbar_kws={"shrink": 0.8})
ax.set_title("Correlation Heatmap of Numeric Features", fontweight="bold", pad=15)
plt.tight_layout()
plt.savefig("viz5_heatmap.png", bbox_inches="tight")
plt.show()
```

### 5.6 Financial Stress and Suicidal Thoughts

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: Financial stress level vs depression ---
sns.countplot(data=df, x="Financial Stress", hue="Depression_Label",
              palette={"Not Depressed": "#5BA4CF", "Depressed": "#E07B7B"},
              ax=axes[0], edgecolor="white")
axes[0].set_title("Financial Stress vs Depression", fontweight="bold")
axes[0].set_xlabel("Financial Stress Level (1=Low, 5=High)")
axes[0].set_ylabel("Count")
axes[0].legend(title="Status")

# --- Right: Suicidal thoughts history vs depression ---
suicidal_data = df.groupby(["Have you ever had suicidal thoughts ?", "Depression_Label"]).size().unstack(fill_value=0)
suicidal_data.plot(kind="bar", ax=axes[1], color=["#E07B7B", "#5BA4CF"], edgecolor="white", width=0.55)
axes[1].set_title("Suicidal Thoughts History vs Depression", fontweight="bold")
axes[1].set_xlabel("Suicidal Thoughts History")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis="x", rotation=0)
axes[1].legend()

plt.tight_layout()
plt.savefig("viz6_stress_suicidal.png", bbox_inches="tight")
plt.show()
```

## 6. Observations and Interpretation

After looking at the data, here are five things we noticed:

**1. More students are depressed than not.**
Out of 27,901 students in the dataset, about 58.5% are depressed and 41.5% are not. This means the data is a little unbalanced — there are more depressed students than non-depressed ones. This is something to keep in mind when building a prediction model later.

**2. Higher academic pressure is linked to more depression.**
Students who rated their academic pressure as 4 or 5 (out of 5) were more likely to be depressed. This makes sense — the more stressed a student is about school, the harder it is to stay mentally healthy.

**3. Students who sleep less tend to be more depressed.**
Students who sleep less than 5 hours a night made up the biggest group of depressed students. On the other hand, students who sleep 7–8 hours tend to have better mental health. Getting enough sleep seems to really matter.

**4. Eating habits and money problems also play a role.**
Students with unhealthy diets had higher depression rates compared to those who eat well. Also, students with higher financial stress (level 4 or 5) were more often depressed. This shows that everyday problems like money and food choices can affect mental health too.

**5. Academic pressure and financial stress are the strongest warning signs.**
Looking at the heatmap, academic pressure and financial stress had the highest connection to depression among all the number columns. Interestingly, CGPA (grades) had a very weak connection — meaning getting good or bad grades alone doesn't really predict if a student is depressed.
