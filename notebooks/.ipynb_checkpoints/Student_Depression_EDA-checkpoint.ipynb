## 1. Setup and Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({"figure.dpi": 120, "axes.titlesize": 13,
                     "axes.labelsize": 11})
print("Libraries loaded successfully.")
```

## 2. Dataset Overview

```python
df = pd.read_csv("Student_Depression_Dataset.csv")
print("First 5 rows:")
df.head()
```

```python
print(f"Shape: {df.shape[0]:,} rows x {df.shape[1]} columns")
print()
print("Column Names and Data Types:")
print(df.dtypes.to_string())
```

## 3. Data Quality Checks

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
quality_df = pd.DataFrame({"Missing Count": missing, "Missing %": missing_pct})
cols_with_missing = quality_df[quality_df["Missing Count"] > 0]
if len(cols_with_missing) > 0:
    print("Columns with missing values:")
    print(cols_with_missing)
else:
    print("No missing values detected.")
    
print(f"\nDuplicate rows: {df.duplicated().sum()}")
```

## 4. Class Distribution (Target Variable)

```python
target_counts = df["Depression"].value_counts().sort_index()
print("Depression Class Distribution:")
for val, count in target_counts.items():
    label = "Depressed" if val == 1 else "Not Depressed"
    pct = count / len(df) * 100
    print(f"  {val} ({label}): {count:,} ({pct:.1f}%)")
```

## 5. EDA Visualizations

### 5.1 Target Class Distribution

```python
labels = ["Not Depressed", "Depressed"]
counts = df["Depression"].value_counts().sort_index().values
colors = ["#5BA4CF", "#E07B7B"]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

bars = axes[0].bar(labels, counts, color=colors, edgecolor="white", linewidth=1.5, width=0.5)
axes[0].set_title("Class Distribution (Bar Chart)", fontweight="bold")
axes[0].set_ylabel("Number of Students")
for bar, count in zip(bars, counts):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 150,
                 f"{count:,}", ha="center", fontsize=11, fontweight="bold")
axes[0].set_ylim(0, max(counts) * 1.15)

wedges, texts, autotexts = axes[1].pie(
    counts, labels=labels, autopct="%1.1f%%", colors=colors,
    startangle=90, textprops={"fontsize": 10})
for at in autotexts:
    at.set_fontweight("bold")
axes[1].set_title("Class Distribution (Pie Chart)", fontweight="bold")

plt.suptitle("Depression Status Among Students", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("viz1_class_distribution.png", bbox_inches="tight")
plt.show()
```

### 5.2 Age Distribution by Depression Status

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for val, label, color in [(0,"Not Depressed","#5BA4CF"), (1,"Depressed","#E07B7B")]:
    axes[0].hist(df[df["Depression"]==val]["Age"], bins=20, alpha=0.65,
                 label=label, color=color, edgecolor="white")
axes[0].set_title("Age Distribution by Depression Status", fontweight="bold")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")
axes[0].legend()

df["Depression_Label"] = df["Depression"].map({0:"Not Depressed", 1:"Depressed"})
sns.boxplot(data=df, x="Depression_Label", y="Age",
            palette={"Not Depressed":"#5BA4CF","Depressed":"#E07B7B"}, ax=axes[1], linewidth=1.5)
axes[1].set_title("Age Boxplot by Depression Status", fontweight="bold")
axes[1].set_xlabel("")
axes[1].set_ylabel("Age")

plt.tight_layout()
plt.savefig("viz2_age_distribution.png", bbox_inches="tight")
plt.show()
```

### 5.3 Academic Pressure and CGPA

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.countplot(data=df, x="Academic Pressure", hue="Depression_Label",
              palette={"Not Depressed":"#5BA4CF","Depressed":"#E07B7B"},
              ax=axes[0], edgecolor="white")
axes[0].set_title("Academic Pressure vs Depression", fontweight="bold")
axes[0].set_xlabel("Academic Pressure (1=Low, 5=High)")
axes[0].set_ylabel("Count")
axes[0].legend(title="Status")

for val, label, color in [(0,"Not Depressed","#5BA4CF"), (1,"Depressed","#E07B7B")]:
    axes[1].hist(df[df["Depression"]==val]["CGPA"], bins=25, alpha=0.65,
                 label=label, color=color, edgecolor="white")
axes[1].set_title("CGPA Distribution by Depression Status", fontweight="bold")
axes[1].set_xlabel("CGPA")
axes[1].set_ylabel("Frequency")
axes[1].legend()

plt.tight_layout()
plt.savefig("viz3_academic.png", bbox_inches="tight")
plt.show()
```

### 5.4 Sleep Duration and Dietary Habits

```python
sleep_order = ["Less than 5 hours","5-6 hours","7-8 hours","More than 8 hours","Others"]
diet_order  = ["Unhealthy","Moderate","Healthy","Others"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sleep_dep = df.groupby(["Sleep Duration","Depression_Label"]).size().unstack(fill_value=0)
sleep_dep = sleep_dep.reindex([s for s in sleep_order if s in sleep_dep.index])
sleep_dep.plot(kind="bar", ax=axes[0], color=["#E07B7B","#5BA4CF"], edgecolor="white", width=0.7)
axes[0].set_title("Sleep Duration vs Depression", fontweight="bold")
axes[0].set_xlabel("Sleep Duration")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis="x", rotation=25)
axes[0].legend()

diet_dep = df.groupby(["Dietary Habits","Depression_Label"]).size().unstack(fill_value=0)
diet_dep = diet_dep.reindex([d for d in diet_order if d in diet_dep.index])
diet_dep.plot(kind="bar", ax=axes[1], color=["#E07B7B","#5BA4CF"], edgecolor="white", width=0.7)
axes[1].set_title("Dietary Habits vs Depression", fontweight="bold")
axes[1].set_xlabel("Dietary Habits")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis="x", rotation=15)
axes[1].legend()

plt.tight_layout()
plt.savefig("viz4_sleep_diet.png", bbox_inches="tight")
plt.show()
```

### 5.5 Correlation Heatmap of Numeric Features

```python
num_cols = ["Age","CGPA","Academic Pressure","Work Pressure",
            "Study Satisfaction","Job Satisfaction","Work/Study Hours",
            "Financial Stress","Depression"]
corr = df[num_cols].corr()

fig, ax = plt.subplots(figsize=(10, 7))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="RdYlBu_r",
            center=0, linewidths=0.5, ax=ax, annot_kws={"size": 9},
            cbar_kws={"shrink": 0.8})
ax.set_title("Correlation Heatmap of Numeric Features", fontweight="bold", pad=15)
plt.tight_layout()
plt.savefig("viz5_heatmap.png", bbox_inches="tight")
plt.show()
```

### 5.6 Financial Stress and Suicidal Thoughts

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.countplot(data=df, x="Financial Stress", hue="Depression_Label",
              palette={"Not Depressed":"#5BA4CF","Depressed":"#E07B7B"},
              ax=axes[0], edgecolor="white")
axes[0].set_title("Financial Stress vs Depression", fontweight="bold")
axes[0].set_xlabel("Financial Stress Level (1=Low, 5=High)")
axes[0].set_ylabel("Count")
axes[0].legend(title="Status")

st_data = df.groupby(["Have you ever had suicidal thoughts ?","Depression_Label"]).size().unstack(fill_value=0)
st_data.plot(kind="bar", ax=axes[1], color=["#E07B7B","#5BA4CF"], edgecolor="white", width=0.55)
axes[1].set_title("Suicidal Thoughts History vs Depression", fontweight="bold")
axes[1].set_xlabel("Suicidal Thoughts History")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis="x", rotation=0)
axes[1].legend()

plt.tight_layout()
plt.savefig("viz6_stress_suicidal.png", bbox_inches="tight")
plt.show()
```

## 6. Observations and Interpretation

The exploratory data analysis of the Student Depression Dataset (n = 27,901) yields several noteworthy findings.

**1. Class Imbalance.** The target variable exhibits a moderate imbalance: approximately 58.5% of respondents are classified as depressed (Depression = 1), compared to 41.5% who are not. While the imbalance is not extreme, it is sufficient to warrant attention during model training — techniques such as stratified sampling or class-weight adjustment may be necessary to prevent the classifier from being biased toward the majority class.

**2. Academic Pressure as a Key Predictor.** Students reporting higher levels of academic pressure (ratings of 4 or 5 out of 5) show a disproportionately higher prevalence of depression. This suggests a strong positive association between perceived academic burden and mental health deterioration, consistent with findings in the broader student mental health literature.

**3. Sleep Deprivation and Depression.** The analysis of sleep duration reveals that students sleeping fewer than five hours per night constitute the largest group of depressed individuals. Conversely, students who sleep seven hours or more tend to exhibit lower rates of depression. This inverse relationship between sleep adequacy and depression risk highlights sleep quality as a potentially actionable protective factor.

**4. Financial Stress and Dietary Habits.** Students experiencing high financial stress (levels 4–5) are more likely to be classified as depressed. Similarly, students with unhealthy dietary habits account for a greater proportion of depressed cases compared to those with healthy or moderate diets. These findings point to socioeconomic and lifestyle factors as significant contributors to student mental health outcomes.

**5. Correlation Structure.** The heatmap indicates that Academic Pressure and Financial Stress share the strongest positive correlations with depression among the numeric features, while Study Satisfaction shows a moderate negative correlation — students who are more satisfied with their academic experience tend to report lower rates of depression. Notably, CGPA shows only a weak correlation with depression, suggesting that academic performance alone is not a reliable proxy for mental well-being.
